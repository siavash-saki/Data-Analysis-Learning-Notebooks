## Number of children of the richest men in the world

```python
import numpy as np
import pandas as pd
```

```python
richest= pd.read_html('https://en.wikipedia.org/wiki/List_of_Americans_by_net_worth')
```

```python
richest_df=richest[0]
richest_df.head()
```

```python
richest_names= list(richest_df['Name'])
richest_names
```

```python
wiki_list=[]
n=-1
for i in richest_names:
    n +=1
    s='https://en.wikipedia.org/wiki/'+i.split()[0]+'_'+i.split()[1]
    try:
        wiki_rich= pd.read_html(s)
        wiki_list.append(wiki_rich[0])
    except:
        wiki_list.append(pd.DataFrame(pd.Series(np.nan)))
        continue
```

```python
children=[]
for i in wiki_list:
    try:
        children.append(i.set_index(i.columns[0]).loc['Children'].iloc[0])
    except:
        children.append(np.nan)
        continue
```

```python
children
```

```python
child=[]
for i in children:
    try: 
        child.append(int(i))
    except:
        try:
            child.append(int(i[0]))
        except:
            child.append(np.nan)
```

```python
child
```

```python
richest_df['Children']=child
```

```python
richest_df
```

```python
richest_df.dropna(inplace=True)
```

```python
richest_df['Children']=richest_df['Children'].astype(int)
```
