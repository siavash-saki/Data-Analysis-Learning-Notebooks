## read and write csv file

```python
df1= pd.read_csv('sales_sample_data.csv')
df1.head()
```

```python
df1_pivot=pd.pivot_table(df1,values='Cost',index=['Seller','Product'],aggfunc=sum).round().astype(int)
df1_pivot
```

```python
df1_pivot.to_csv('sales_per_person_sample.csv')
```
